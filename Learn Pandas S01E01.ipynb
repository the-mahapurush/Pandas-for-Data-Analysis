### To get a particular column value of multiple rows in dataframe <br> With `iloc` we can pass only integer but with `loc` we can pass *labels* too

```python
dataset_bbc.iloc[[0,1,2,3],0] #Title of first 4 Rows
```

```python
dataset_bbc.loc[[0,1,2,3],'link']
#We can pass label in loc but with iloc we can pass only integer
```

```python

dataset_bbc.loc[[30,1545,2455],['title','pubDate']]
```
