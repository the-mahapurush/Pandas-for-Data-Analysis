### Sactterplot
#### A scatter plot is used when you want to visualize the relationship between two continuous variables. It helps you to see how one variable is affected by another or if there is any correlation between the two.
Some common scenarios for using scatter plots include:
<ol>
<li>Identifying correlations: Whether the relationship is positive, negative, or none at all.</li>
<li>Detecting outliers: Values that are significantly different from the rest of the data points.</li>
<li>Trend analysis: Showing how one variable changes as another variable changes.</li>
<li>Visualizing distributions: Especially when combined with other plots or line fits to understand the spread and pattern.</li>
</ol>

A pie chart is a circular graph divided into segments (or "slices") to illustrate numerical proportions. ## STEM PLOT
A stem plot in Matplotlib is a type of plot where vertical lines (or "stems") are drawn from a baseline (usually the x-axis) to the data points. It's a good way to show discrete data or emphasize individual data points while also showing their distribution.

Stem plots are particularly useful for visualizing:

>Discrete data points<br>
Showing small-scale variations in data <br>
Displaying time-series data where individual data points matter more than the overall trend <br>

Stem Plot: Better for emphasizing individual data points and showing their exact distribution.

Bar Chart: Better for comparing quantities across different categories or groups.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = np.array([2, 5, 8, 3, 10])
```

```python
# Create stem plot
plt.stem(x, y)
plt.title('Basic Stem Plot')
plt.show()
```

```python
plt.stem(x, y,linefmt=":")
plt.title('Basic Stem Plot')
plt.show()
```

### STEM vs BAR
Lets use the following data:

Categories: A, B, C, D, E
Values: 5, 10, 15, 20, 25

We'll compare how the stem plot and the bar chart represent this data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for stem plot
x = np.arange(1, 6)  # X values (A, B, C, D, E)
y = np.array([5, 10, 15, 20, 25])  # Y values (5, 10, 15, 20, 25)

# Create stem plot
plt.stem(x, y)
plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])  # Set custom labels for x-axis
plt.title("Stem Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for bar chart
x = np.arange(1, 6)  # X values (A, B, C, D, E)
y = np.array([5, 10, 15, 20, 25])  # Y values (5, 10, 15, 20, 25)

# Create bar chart
plt.bar(x, y)
plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])  # Set custom labels for x-axis
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

<ul>
   <li> Stem Plot: You will see vertical lines (stems) at each category (A, B, C, D, E) with a marker at the top of each line. This plot emphasizes individual data points and their exact values. </li>
<li>Bar Chart: You will see solid bars, with the height of each bar corresponding to the values (5, 10, 15, 20, 25). This plot helps to visually compare the magnitude of the values across different categories.</li>
</ul>

##### Stem Plot emphasizes individual data points, showing both the data and its distribution, with clear visibility of the exact value of each data point.
##### Bar Chart is better for comparing the relative magnitudes of data across different categories, where each bar represents a group.

## BOX PLOT

A box plot (also called a box-and-whisker plot) is a graphical representation of the distribution of a dataset. It shows the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum of the data. It is particularly useful for identifying outliers, understanding the spread, and comparing distributions between multiple datasets.

<b>Key Elements of a Box Plot<b>
<ol>
<li>Minimum: The smallest data point (excluding outliers).</li>
<li>First Quartile (Q1): The median of the lower half of the data.</li>
<li>Median (Q2): The middle data point (50th percentile).</li>
<li>Third Quartile (Q3): The median of the upper half of the data.</li>
<li>Maximum: The largest data point (excluding outliers).</li>
<li>Whiskers: Lines extending from the box that represent the range of data, excluding outliers.</li>
<li>Outliers: Data points that fall outside the "whiskers" (often represented as dots).</li>
</ol>

#### Syntax
x=[ ] <br>
plt.boxplot(x) <br>
plt.show

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
data = np.random.normal(0, 100, 100)  # 100 points, normally distributed with mean 0 and std 1

# Create a box plot
plt.boxplot(data)

# Add title and labels
plt.title("Box Plot Example")
plt.ylabel("Values")

# Display the plot
plt.show()
```

```python
x=[10,20,30,40,50]
plt.boxplot(x,notch=False,vert=True,whis=3,showmeans=True)
plt.show()
```

## STACK PLOT / AREA PLOT
### A stack plot is used to visualize the distribution of multiple datasets over time or across different categories. It shows how the values of different categories or components accumulate and contribute to the total value at each point. Each dataset is stacked on top of the previous one, allowing you to see how individual components change over time.

x=[ ] <br>
y=[ ] <br>
plt.stackplot(x,y)

```python
import matplotlib.pyplot as plt
import numpy as np

# Time points (x values)
x = np.arange(0, 10)

# Data for each category (y values for each category over time)
y1 = np.random.rand(10)
y2 = np.random.rand(10)
y3 = np.random.rand(10)

# Stack plot
plt.stackplot(x, y1, y2, y3, labels=['Category 1', 'Category 2', 'Category 3'], alpha=0.6)
# alpha for opacity
# Adding labels and title
plt.title('Stack Plot Example')
plt.xlabel('Time')
plt.ylabel('Values')

# Show legend
plt.legend(loc='upper left')
# a legend is a key component used to label and identify different elements (like lines, bars, or other plot elements) within a plot.
#It helps to make the plot more understandable by showing what each plot element represents, especially when the plot contains multiple datasets or groups.
# Display the plot
plt.show()
```

### Understanding legend

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot lines
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')

# Add a legend
plt.legend()

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot with Legend')

# Display the plot
plt.show()
```

### Customizing Legend

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot lines
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='--', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='-', linewidth=2)

# Add a customized legend
plt.legend(loc='lower left', fontsize=10, frameon=False)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Customized Line Plot with Legend')

# Display the plot
plt.show()
```

### Scenario: "The Life of a Cat" 

Let's imagine a day in the life of a cat. We'll track how much time the cat spends on various activities during the day:

- **Sleeping**: Cats are known for sleeping a lot.
- **Eating**: Cats love to eat!
- **Playing**: A cat's playtime is essential for keeping it entertained.
- **Staring Out of the Window**: Sometimes cats just stare at nothing outside for hours.
- **Begging for Attention**: Cats tend to demand attention at random moments.


```python
import matplotlib.pyplot as plt
import numpy as np

# Time points (representing hours of the day)
x = np.arange(1, 25)

# Data for each category (hours spent on each activity)
sleeping = np.random.rand(24) * 12  # Cats sleep a lot (up to 12 hours)
eating = np.random.rand(24) * 2  # Cats eat, but not too much
playing = np.random.rand(24) * 3  # Playtime varies, not too much
staring_window = np.random.rand(24) * 4  # Cats stare out of the window
begging_attention = np.random.rand(24) * 2  # Random begging for attention

# Stack plot with cat activities
plt.stackplot(x, sleeping, eating, playing, staring_window, begging_attention, 
              labels=['Sleeping', 'Eating', 'Playing', 'Staring out the window', 'Begging for attention'],
              alpha=0.9)

# Adding labels and title
plt.title('A Day in the Life of a Cat')
plt.xlabel('Hours of the Day')
plt.ylabel('Time Spent (in hours)')
plt.xticks(np.arange(1, 25, 3))  # Set x-ticks every 3 hours for readability

# Add a fun legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

### Well track how a familys monthly expenses are distributed among different categories:

-> Rent <br>
-> Groceries <br>
-> Transportation <br>
-> Utilities <br>
-> Entertainment

```python
import matplotlib.pyplot as plt
import numpy as np

# Time points (representing each day of the month)
days = np.arange(1, 31)

# Example data: Expenses (in dollars) for each category on each day of the month
rent = np.full(30, 500)  # Rent is constant throughout the month
groceries = np.random.randint(10, 50, 30)  # Random spending on groceries
transportation = np.random.randint(5, 30, 30)  # Random transportation costs
utilities = np.random.randint(20, 100, 30)  # Random utility bills
entertainment = np.random.randint(0, 100, 30)  # Random entertainment costs

# Create the stack plot
plt.stackplot(days, rent, groceries, transportation, utilities, entertainment,
              labels=['Rent', 'Groceries', 'Transportation', 'Utilities', 'Entertainment'],
              alpha=0.7)

# Add titles and labels
plt.title('Monthly Expenses of a Family')
plt.xlabel('Days of the Month')
plt.ylabel('Amount Spent ($)')
plt.xticks(np.arange(1, 31, 5))  # Show x-ticks for every 5th day for better readability

# Add a legend to identify each category
plt.legend(loc='upper left')

# Display the plot
plt.show()
```

### Explanation of the Plot:

#### X-axis (Days of the Month):
- The **X-axis** represents the **days of the month** (from 1 to 30 days).
- Each tick on the **X-axis** corresponds to a specific day.

#### Y-axis (Amount Spent in Dollars):
- The **Y-axis** represents the **total amount of money spent** each day (in dollars).
- The range on the **Y-axis** would scale according to the highest value in any category (in our case, its dependent on the random values generated for each category).

#### Stacked Categories:
- The different colored blocks represent the different **expense categories**:
  - **Rent**: Fixed amount of $500 per day, usually just a one-time payment each month, but represented consistently for the entire month here.
  - **Groceries**: Random daily spending between $10 and $50.
  - **Transportation**: Random daily transportation costs between $5 and $30.
  - **Utilities**: Random utility costs between $20 and $100.
  - **Entertainment**: Random entertainment spending between $0 and $100.

#### Interpreting the Stack:
- Each stacked segment shows how much is being spent on a particular category on a given day.
- For example, if **Day 10** shows a larger amount in the **Groceries** category, you can tell that this day had a bigger grocery purchase.
- The **height of the stack** on any given day represents the **total amount spent on all categories** combined for that day.

#### Insights:
- By looking at this stack plot, we can quickly see how much the family spends on different categories, and where their money is going over the month.
- The **Rent** will likely dominate the plot (if we keep the rent amount large), but other categories like **Entertainment** and **Utilities** will show up as smaller, more varied amounts over the course of the month.
- You might notice that **Groceries** and **Transportation** have a more random, fluctuating pattern.

#### Why Use a Stack Plot for This?
- **Stack plots** are helpful when you want to see the **total contribution of different categories** to a whole.
- In this case, we can see how all the expenses stack up together over the course of a month, and how much each category contributes on any given day.
